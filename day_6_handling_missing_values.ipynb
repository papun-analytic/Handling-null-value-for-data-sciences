### If the missing value isn’t identified as NaN , then we have to first convert or replace such non NaN entry with a NaN.

### filling missing value with mean,mode and median

```python
data5["LoanAmount"]=data5["LoanAmount"].fillna(np.mean)
```

```python
data5["Gender"]=data5["Gender"].fillna(pd.Series.mode)
```

```python
data5['Credit_History']=data5["Credit_History"].fillna(np.median)
```

```python
data5.isnull().sum()
```

###  filling value with backward and forward
##### backward or  forward filling is used for time series data analysis

```python
data5["Married"]=data5["Married"].fillna(method='ffill')
```

```python
data5["Dependents"]=data5["Dependents"].fillna(method="bfill")
```

```python
data5.isnull().sum()
```

### groupby imputation
#### left side should be categorical and right should be numerical
#### left side categorical value should not contain any null value

```python
data6=pd.read_csv("Loan_data.csv")
```

```python
data6.isnull().sum()
```

```python
data6["LoanAmount"]=data6.groupby(["Property_Area"])["LoanAmount"].transform(lambda x: x.fillna(x.mean()))
```

```python
data6.isnull().sum()
```

### liner interpolation technique

```python
data6.interpolate(method ='linear', limit_direction ='forward')
```

```python
data6.isnull().sum()
```

```python
data6.interpolate(method ='linear', limit_direction ='backward', limit = 1)
```

```python
data6.isnull().sum()
```
